# Locomotion Mode CNN Classifier

**Goal:** Classify locomotion mode (walk / stair ascent / stair descent / ramp ascent / ramp descent) from IMU + joint angle time series.

**Data:** CAMARGO dataset (prototype) → swap to WAWA data later.

**Architecture:** Lightweight 1D CNN suitable for microcontroller deployment.

**Input:** 10 channels × window_size samples
- 6 channels: thigh IMU (accel_x/y/z, gyro_x/y/z)
- 2 channels: hip angles (L/R) from goniometer (proxy for motor encoder)
- 2 channels: hip angular velocities (L/R, numerically differentiated)

**Output:** 5-class softmax (walk, stair_ascent, stair_descent, ramp_ascent, ramp_descent)

## 3. Generic Windowing (works with any data source)

This part is data-source agnostic — once you have a (time × channels) array and labels, this creates training windows.

```python
def create_windows(data, label, window_size=WINDOW_SIZE, stride=WINDOW_STRIDE):
    """
    Slice a (T, C) array into (N, C, W) windows with a single label.
    
    Args:
        data: np.array of shape (T, C) — T timesteps, C channels
        label: int — class label for all windows
        window_size: int — samples per window
        stride: int — step between windows
    
    Returns:
        X: np.array of shape (N, C, W)
        y: np.array of shape (N,)
    """
    T, C = data.shape
    windows = []
    for start in range(0, T - window_size + 1, stride):
        window = data[start:start + window_size, :]  # (W, C)
        windows.append(window.T)  # (C, W) — channels first for Conv1d
    
    if len(windows) == 0:
        return np.empty((0, C, window_size)), np.empty(0, dtype=int)
    
    X = np.array(windows, dtype=np.float32)
    y = np.full(len(windows), label, dtype=np.int64)
    return X, y


def normalize_channels(X_train, X_test):
    """
    Z-score normalize each channel using training statistics.
    X shape: (N, C, W)
    """
    N, C, W = X_train.shape
    # Compute mean/std per channel across all training windows and timesteps
    mean = X_train.reshape(-1, C, W).mean(axis=(0, 2), keepdims=True)  # (1, C, 1)
    std = X_train.reshape(-1, C, W).std(axis=(0, 2), keepdims=True) + 1e-8
    
    X_train_norm = (X_train - mean) / std
    X_test_norm = (X_test - mean) / std
    return X_train_norm, X_test_norm, mean, std


# Quick test with synthetic data
dummy_data = np.random.randn(500, N_CHANNELS).astype(np.float32)
X_dummy, y_dummy = create_windows(dummy_data, label=0)
print(f'Synthetic test: {X_dummy.shape} windows, labels {y_dummy.shape}')
print(f'  Window shape: ({X_dummy.shape[1]} channels, {X_dummy.shape[2]} timesteps)')
```

## 4. CNN Model

Lightweight 1D CNN — 3 conv layers with batch norm, global average pooling, single FC layer.

**Design goals:**
- Small enough for Raspberry Pi / microcontroller (~50K parameters)
- Fast inference (<2ms per window)
- Comparable to literature architectures achieving ~94% on similar tasks

```python
class LocomotionCNN(nn.Module):
    """
    Lightweight 1D CNN for locomotion mode classification.
    
    Input:  (batch, n_channels, window_size)
    Output: (batch, n_classes) — logits
    """
    def __init__(self, n_channels=N_CHANNELS, n_classes=N_CLASSES, 
                 filters=(32, 64, 64), kernel_size=5, dropout=0.3):
        super().__init__()
        
        layers = []
        in_ch = n_channels
        for out_ch in filters:
            layers.extend([
                nn.Conv1d(in_ch, out_ch, kernel_size, padding=kernel_size//2),
                nn.BatchNorm1d(out_ch),
                nn.ReLU(inplace=True),
            ])
            in_ch = out_ch
        
        self.features = nn.Sequential(*layers)
        self.pool = nn.AdaptiveAvgPool1d(1)  # global average pooling
        self.dropout = nn.Dropout(dropout)
        self.classifier = nn.Linear(filters[-1], n_classes)
    
    def forward(self, x):
        x = self.features(x)     # (B, C_out, W)
        x = self.pool(x)         # (B, C_out, 1)
        x = x.squeeze(-1)        # (B, C_out)
        x = self.dropout(x)
        x = self.classifier(x)   # (B, n_classes)
        return x


# Instantiate and check
model = LocomotionCNN().to(device)
n_params = sum(p.numel() for p in model.parameters())
print(f'Model parameters: {n_params:,}')
print(f'Architecture:\n{model}')

# Test forward pass
x_test = torch.randn(4, N_CHANNELS, WINDOW_SIZE).to(device)
out = model(x_test)
print(f'\nForward pass: input {x_test.shape} → output {out.shape}')
```

## 5. Training Loop

```python
def train_epoch(model, loader, criterion, optimizer):
    model.train()
    total_loss, correct, total = 0, 0, 0
    for X_batch, y_batch in loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        logits = model(X_batch)
        loss = criterion(logits, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * len(y_batch)
        correct += (logits.argmax(1) == y_batch).sum().item()
        total += len(y_batch)
    return total_loss / total, correct / total


@torch.no_grad()
def eval_epoch(model, loader, criterion):
    model.eval()
    total_loss, correct, total = 0, 0, 0
    all_preds, all_labels = [], []
    for X_batch, y_batch in loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        logits = model(X_batch)
        loss = criterion(logits, y_batch)
        total_loss += loss.item() * len(y_batch)
        preds = logits.argmax(1)
        correct += (preds == y_batch).sum().item()
        total += len(y_batch)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(y_batch.cpu().numpy())
    return total_loss / total, correct / total, all_preds, all_labels


def train_model(model, X_train, y_train, X_val, y_val,
                epochs=50, batch_size=64, lr=1e-3):
    """
    Full training pipeline with early stopping.
    X shape: (N, C, W), y shape: (N,)
    """
    train_ds = TensorDataset(torch.from_numpy(X_train), torch.from_numpy(y_train))
    val_ds = TensorDataset(torch.from_numpy(X_val), torch.from_numpy(y_val))
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=batch_size)
    
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, patience=5, factor=0.5, verbose=True)
    
    best_val_acc = 0
    best_state = None
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    
    for epoch in range(epochs):
        train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer)
        val_loss, val_acc, _, _ = eval_epoch(model, val_loader, criterion)
        scheduler.step(val_loss)
        
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
        
        if (epoch + 1) % 5 == 0 or epoch == 0:
            print(f'Epoch {epoch+1:3d}/{epochs}  '
                  f'train_loss={train_loss:.4f}  train_acc={train_acc:.3f}  '
                  f'val_loss={val_loss:.4f}  val_acc={val_acc:.3f}')
    
    # Restore best model
    if best_state:
        model.load_state_dict(best_state)
    print(f'\nBest val accuracy: {best_val_acc:.3f}')
    return history
```

## 6. Run Training (with real or synthetic data)

Once the CAMARGO data loader is working, replace the synthetic data below.

```python
# ============================================================
# SYNTHETIC DATA — replace with real CAMARGO data
# ============================================================
print('Using synthetic data for pipeline test...')
np.random.seed(42)

n_samples_per_class = 200
X_all, y_all = [], []
for cls in range(N_CLASSES):
    # Each class gets slightly different signal characteristics
    freq = 2.0 + cls * 0.5
    t = np.linspace(0, WINDOW_SIZE / SAMPLE_RATE, WINDOW_SIZE)
    for _ in range(n_samples_per_class):
        signal = np.zeros((N_CHANNELS, WINDOW_SIZE), dtype=np.float32)
        for ch in range(N_CHANNELS):
            phase = np.random.uniform(0, 2 * np.pi)
            signal[ch] = (np.sin(2 * np.pi * freq * t + phase) * (1 + 0.2 * ch)
                         + np.random.randn(WINDOW_SIZE) * 0.3)
        X_all.append(signal)
        y_all.append(cls)

X_all = np.array(X_all, dtype=np.float32)
y_all = np.array(y_all, dtype=np.int64)

# Shuffle and split 80/20
perm = np.random.permutation(len(X_all))
X_all, y_all = X_all[perm], y_all[perm]
split = int(0.8 * len(X_all))
X_train, X_val = X_all[:split], X_all[split:]
y_train, y_val = y_all[:split], y_all[split:]

# Normalize
X_train, X_val, _, _ = normalize_channels(X_train, X_val)

print(f'Train: {X_train.shape}, Val: {X_val.shape}')
print(f'Class distribution (train): {Counter(y_train.tolist())}')

# Train
model = LocomotionCNN().to(device)
history = train_model(model, X_train, y_train, X_val, y_val, epochs=30)
```

## 7. Evaluation

```python
# Training curves
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history['train_loss'], label='Train')
ax1.plot(history['val_loss'], label='Val')
ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss')
ax1.set_title('Loss'); ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(history['train_acc'], label='Train')
ax2.plot(history['val_acc'], label='Val')
ax2.set_xlabel('Epoch'); ax2.set_ylabel('Accuracy')
ax2.set_title('Accuracy'); ax2.legend(); ax2.grid(True, alpha=0.3)

fig.tight_layout()
plt.show()
```

```python
# Confusion matrix
val_ds = TensorDataset(torch.from_numpy(X_val), torch.from_numpy(y_val))
val_loader = DataLoader(val_ds, batch_size=64)
_, val_acc, preds, labels = eval_epoch(
    model, val_loader, nn.CrossEntropyLoss())

print(f'Validation accuracy: {val_acc:.3f}\n')
print(classification_report(labels, preds, target_names=LABEL_NAMES))

cm = confusion_matrix(labels, preds)
fig, ax = plt.subplots(figsize=(7, 6))
disp = ConfusionMatrixDisplay(cm, display_labels=LABEL_NAMES)
disp.plot(ax=ax, cmap='Blues', values_format='d')
ax.set_title('Confusion Matrix', fontweight='bold')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

## 8. Inference Speed Test

```python
import time

model.eval()
x_single = torch.randn(1, N_CHANNELS, WINDOW_SIZE).to(device)

# Warmup
for _ in range(100):
    _ = model(x_single)

# Benchmark
if device.type == 'cuda':
    torch.cuda.synchronize()
t0 = time.perf_counter()
n_iters = 1000
for _ in range(n_iters):
    _ = model(x_single)
if device.type == 'cuda':
    torch.cuda.synchronize()
elapsed = time.perf_counter() - t0

print(f'Inference time: {elapsed/n_iters*1000:.2f} ms per window')
print(f'Throughput: {n_iters/elapsed:.0f} windows/sec')
print(f'\nFor 100Hz control loop: need <10ms per inference → '
      f'{"OK" if elapsed/n_iters < 0.01 else "TOO SLOW"}')
```
